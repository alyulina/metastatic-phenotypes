```python
import os
import re
import time
import importlib
import math
import random
import pandas as pd
import numpy as np
import scipy as sp
from scipy import stats
import itertools
from itertools import combinations
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib import cm
import matplotlib.font_manager

import utils 
importlib.reload(utils); 
utils.versions()

metadata = utils.metadata 
clIDs = utils.clIDs
path_to_reads = utils.path_to_reads 
path_to_bootstraps = utils.path_to_bootstraps
```

### First series of experiments

```python
# 3w F1 iv lung samples from exp. one
what = ['1', 20, 'C57B6/129S F1', 'intravenous', 'lung']
samples = list(set(metadata.loc[(metadata['experiment'] == what[0]) & 
                                (metadata['time point, d'] > what[1]) & # be careful here
                                (metadata['genotype'] == what[2]) &
                                (metadata['injection method'] == what[3]) & 
                                (metadata['tissue'] == what[4])
].apply(lambda row: row['notes'].split()[0] if row['notes'] != 'none' else row.name, axis=1).tolist())) # choosing the correct name for merged counts
tag = '_'.join([str(w).replace('/', '').replace(' ', '') for w in what])


start = time.time(); print(f'{len(samples)} mice in the group.')
clID__relative_burden_bootstraps, _, clID__sizes_stats_bootstraps = utils.bootstrap_burden_n_sizes(samples, B_mice=100, B_tumors=100, weighted=True)
_, clID__n_tumors_bootstraps, _ = utils.bootstrap_burden_n_sizes(samples, B_mice=100, B_tumors=100, weighted=False) # do not weight mice

np.save(os.path.join(path_to_bootstraps, f'burden_bootstrap__{tag}.npy'), np.array(list(clID__relative_burden_bootstraps.items()), dtype=object))
np.save(os.path.join(path_to_bootstraps, f'n_tumors_bootstrap__{tag}.npy'), np.array(list(clID__n_tumors_bootstraps.items()), dtype=object))
np.save(os.path.join(path_to_bootstraps, f'sizes_bootstrap__{tag}.npy'), np.array(list(clID__sizes_stats_bootstraps.items()), dtype=object))

end = time.time(); print(f'Done! Took {end - start:.2f} seconds.')
```
